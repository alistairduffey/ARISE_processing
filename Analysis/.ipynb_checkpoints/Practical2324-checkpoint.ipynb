In this practical we will analyze the climate response to stratospheric aerosol geoengineering using results from the Assessing Responses and Impacts of Solar climate intervention on the Earth system with stratospheric aerosol injection (ARISE-SAI) runs. You will need to write about these ARISE-SAI experiments in the methods section of your group report. You can find out more about them here: https://gmd.copernicus.org/articles/15/8221/2022/. This paper describes the experiment design, and some initial results from the ARISE simulations carried out in one climate model (CESM2). We will be using data from the same experimental set-up, carried out in another model, the UK's climate mode: UKESM1. The paper describing the ARISE scenario as carried out in UKESM is here: https://acp.copernicus.org/articles/23/13369/2023/acp-23-13369-2023.html


The ARISE scenario is desinged to hold global mean temperature at 1.5°C above pre-industrial from the year 2035, using injection at multiple latitudes controlled by an algorithm. This algorithm doesn't just set the overall magnitude of injection, but also designs a spatial pattern in an attempt to control the spatial pattern of cooling. The algorithm targets not only global mean temperature, but also the equator-to-pole temperature gradient and the interhemispheric temperature difference - see the papers above for more detail.


In order to assess the impact of SAI, we need to compare against a control case, or 'baseline'. We will compare against the same background GHG emissions scenario (SSP2-4.5) under which the SAI is implemented in ARISE. 


We'll be comparing 3 different cases:

* ssp245_baseline: the climate at 1.5°C under SSP2-4.5, 2013 - 2022 in this climate model. N.B - UKESM is a 'hot' model, with a high cliamte sensitivy, which is why this period is in the past, despite the world only having warmed ~1.2°C to date in reality. 
* ssp245: A warmer, future world under this medium-high emissions scenario. 2050-2069. 
* ARISE: The warmer world, but now with the addition of stratopsheric aerosol injection, over the same period 2050-2069.

You can find a more general overview of stratospheric aerosol geoengineering here: https://onlinelibrary.wiley.com/doi/full/10.1002/wcc.423. This will be useful for the introduction to your report.

- **ssp245**: the late 21st century in our global warming scenario: 2050 - 2069 in the control experiment. This is the future under moderate/high GHG emissions and global warming.

First, I'll plot the temperature change between the global warming case and the baseline. Then you will do the same for the temperature change between the solar_geo case and the baseline.

```python
# let's define a function to avoid typing out lots of complicated file paths.

# Note: the values given here are taken by python as the default values ..
# .. so if you call the function without specificying a season.. 
# ..python will assume you want 'all', i.e. the annual mean. 

def get_data(scenario, season='all', type='mean'):
    # scenario: 'SSP245', 'SSP245_baseline', 'ARISE', or 'preindustrial'
    # season: 'all', 'DJF', 'MAM', 'JJA', or 'SON'
    # type: 'mean' or 'std'
    
    path = '../Output_data/{a}/{a}_{b}_{c}.nc'.format(a=scenario, b=season, c=type)
    data = xr.open_dataset(path)
    return data
```

```python
# As before, we define a variable to store the time-mean output and give it a name that makes sense.
ssp245 = get_data('SSP245', 'all', 'mean')
baseline = get_data('SSP245_baseline', 'all', 'mean')

# We can directly plot the anomaly by subtracting one from the other
# xarray can make a quick simple plot by calling .plot() on a dataarray
# note that we need to specify the 'tas' variable here, because ssp245 and baseline are datasets containing multiple variables
(ssp245 - baseline)['tas'].plot()
```

# Bonus problem:

You've probably noticed the annoying white line along the Greenwich meridian (i.e. 0 longitude) in the plots. This is a common issue with plotting data that crosses the 0 line as a filled contour plot. It's not a major issue scientifically, but it is annoying. The way to solve it is to add a 'cyclic point' which bridges the gap. 

Helpfully, our mapping package cartopy already has a function to deal with this problem:

```python
from cartopy.util import add_cyclic_point
```

We use this package to generate two things: (1) a new data set with an extra data point interpolated at the meridian, and (2) a new set of longitude coordinates with the extra point. The function returns both things when called, so you need to get them in one line. Note that we need to run this function on a dataarray, not a dataset. For example:

```python
ARISE_tas_cyclic_point, lons_cyclic_point = add_cyclic_point(ARISE.tas, coord=ARISE.x)
```

Try using this function to repeat a plot from above, without the white line:

```python
#####################################
######### SOLUTION ##################
#####################################

data_to_plot_1 = (ssp245 - baseline)['tasmax']
data_to_plot_2 = (ARISE - baseline)['tasmax']

lats1 = data_to_plot_1.y
lats2 = data_to_plot_2.y
lats = [lats1, lats2]

data_to_plot_1, lons1 = add_cyclic_point(data_to_plot_1, data_to_plot_1.x)
data_to_plot_2, lons2 = add_cyclic_point(data_to_plot_2, data_to_plot_2.x)
lons = [lons1, lons2]

scenarios = ['SSP2-4.5', 'ARISE-SAI']
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(12,8), 
                        subplot_kw={'projection': ccrs.PlateCarree()})

i=0
for d in [data_to_plot_1, data_to_plot_2]:    
    p = axs[i].contourf(lons[i], lats[i], d, 
                 transform=ccrs.PlateCarree(),
                 cmap='bwr',
                 levels = np.arange(-5, 6, 1),
                  extend='both'      
                         )

    axs[i].coastlines()
    axs[i].set_title('Temperature difference: {} - baseline'.format(scenarios[i]))
    i=i+1


### Add a colorbar
# Adjust the layout to make room for the colorbar on the right
plt.subplots_adjust(right=0.8, hspace=0.4)
# Create a colorbar to the right of the subplots
cax = fig.add_axes([0.85, 0.15, 0.05, 0.7])  # (left, bottom, width, height)
cbar = plt.colorbar(p, cax=cax, orientation='vertical', label='°C')

plt.tight_layout() # this call tightens up the arrangement of subplots to lose white space, often useful. 
plt.show()
```

# 6. Regional plots

So far we've only plotted world maps. Plotting a resticted region is pretty simple, we just include the coordinates of the boundaries in a line as follows:

    ax.set_extent([-180, 180, -30, 30], ccrs.PlateCarree())

the last part of this specifies the map projection (here a regular lat/lon). 

The code below shows this in action:

```python
#####################################
######### SOLUTION ##################
#####################################

data_to_plot = 100*((ssp245 - baseline)/baseline)['pr']
lats = data_to_plot.y


data_to_plot, lons = add_cyclic_point(data_to_plot, data_to_plot.x)

scenario = 'ARISE-SAI'
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,6), 
                        subplot_kw={'projection': ccrs.PlateCarree()})

p = ax.contourf(lons, lats, data_to_plot, 
                     transform=ccrs.PlateCarree(),
                     cmap='BrBG',
                     levels = np.arange(-10, 12, 2),
                      extend='both'      
                             )
ax.set_extent([-7, 3, 49, 60], ccrs.PlateCarree())
ax.coastlines()
ax.set_title('Precipitation difference: {} - baseline (%)'.format(scenario))


### Add a colorbar
# Adjust the layout to make room for the colorbar on the right
plt.tight_layout() # this call tightens up the arrangement of subplots to lose white space, often useful. 

plt.subplots_adjust(right=0.8, hspace=0.4)
# Create a colorbar to the right of the subplots
cax = fig.add_axes([0.85, 0.15, 0.05, 0.7])  # (left, bottom, width, height)
cbar = plt.colorbar(p, cax=cax, orientation='vertical', label='%')

plt.show()
```

```python
# lets define a function for an area weighted mean:
# in x-array, this is as simple as doing .weighted(weights):
def area_weighted_mean(data):   
    weighted_data = data.weighted(grid_cell_areas.areacella)
    weighted_mean = weighted_data.mean()
    return weighted_mean
```
