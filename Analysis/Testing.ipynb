```python
import xarray as xr
import numpy as np
```

```python
season = 'all' #'DJF' 
var='siconca'

data_path = '../Output_data/'
arise = xr.open_dataset(data_path+'ARISE/ARISE_{}_mean.nc'.format(season))
ssp245 = xr.open_dataset(data_path+'SSP245/SSP245_{}_mean.nc'.format(season))
baseline = xr.open_dataset(data_path+'SSP245_baseline/SSP245_baseline_{}_mean.nc'.format(season))
preindustrial = xr.open_dataset(data_path+'preindustrial/preindustrial_{}_mean.nc'.format(season))
```

```python
(arise-ssp245)[var].plot()
```

```python
(arise-baseline)[var].plot()
```

```python
(100*(arise-baseline)/(baseline-preindustrial))[var].plot(levels=np.arange(-100, 100, 20))
```
